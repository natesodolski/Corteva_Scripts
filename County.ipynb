```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\nasod\Downloads\Corn Yield by County 1940-2023.csv')
```

```python
df = df[df['Year'].isin([1943,1945,1950,1958,1979,1992,1996,2008,2011,2015])]
```

```python
df_indiana = df[df['State'].str.strip() == 'INDIANA']
```

```python
df_indiana
```

```python
from IPython.display import FileLink
df_indiana.to_csv('df_indiana.csv', index=False)
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
corn_yield_df = df_indiana[df_indiana['Data Item'] == 'CORN, GRAIN - YIELD, MEASURED IN BU / ACRE']

plt.figure(figsize=(10, 12))
plt.barh(corn_yield_df['County'], corn_yield_df['Value'], color='blue')
plt.xlabel('Yield (BU / ACRE)')
plt.title('Corn Grain Yield per Acre by County  (2024)')
plt.tight_layout()

plt.show
```

```python
import seaborn as sns

plt.figure(figsize=(18, 6))
sns.boxplot(x='County', y='CV (%)', data=df_indiana)
plt.xlabel('Indiana Counties')
plt.ylabel('CV %')
plt.title('CV % Distriibution by Indiana Counties')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the plot
plt.show()
```
